```python
import pandas as pd
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
```

```python
df = pd.read_csv('../csv/dados_vendas.csv')
```

```python
df['Numero_Indice_Total'] = df['Numero_Indice_Total'].str.replace(',','.')
df['Numero_Indice_Total'] = df['Numero_Indice_Total'].astype('float')
df['Var. mensal (%)Total'] = df['Var. mensal (%)Total'].str.replace('%','')
df['Var. mensal (%)Total'] = df['Var. mensal (%)Total'].str.replace(',','.')
df['Var. mensal (%)Total'] = df['Var. mensal (%)Total'].astype('float') / 100
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].str.replace(',','.')
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].str.replace('%','')

df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].astype('float')/100
df['Data'] = pd.to_datetime(df['Data'],format='%b/%y')
```

```python
df.replace(".", None, inplace=True)
```

```python
df.to_csv('../data/dados_vendas_joao_pessoa_tratados.csv', index=False, sep=';', encoding='utf-8')
```
