```python
df['Número-Índice Total'] = df['Número-Índice Total'].str.replace(',','.')
```

```python
df
```

```python
df['Número-Índice Total'] = df['Número-Índice Total'].astype('float')
```

```python
df['Var. Mensal (%) Total'] = df['Var. Mensal (%) Total'].str.replace('%','')
```

```python
df['Var. Mensal (%) Total'] = df['Var. Mensal (%) Total'].str.replace(',','.')
```

```python
df['Var. Mensal (%) Total'] = df['Var. Mensal (%) Total'].astype('float') / 100
```

```python
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].str.replace(',','.')
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].str.replace('%','')
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].astype('float')
```

```python
df['Var. em 12 meses (%)Total'] = df['Var. em 12 meses (%)Total'].astype('float')/100
```

```python
df
```

```python
df['Data'] = pd.to_datetime(df['Data'],format='%b/%y')
```

```python
df.to_csv('../data/dados_vendas_tratados.csv', index=False, sep=';', encoding='utf-8')
```
